#Step By Step Guide

1. Import Libraries
First, import the necessary libraries in your Python script.

```python
from bs4 import BeautifulSoup
import requests
```

2. Make a Request to the Web Page
Use the Requests library to make an HTTP request to the web page you want to scrape.

```python
url = 'https://www.ekomi.fr/avis-clients-copy-top.html'
response = requests.get(url)
```

3. Parse content with beautifulSoup

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

Identify the HTML elements that contain the reviews. You might need to inspect the web page's source code to understand its structure. For example, if reviews are in <div> tags with a class review, find them like this:

```python
reviews = soup.find_all('div', class_='review-item-body review-coments')
```

```python
print(reviews)
```

```python
print(soup)
```

5. Extract Review Text
Loop through the located elements to extract the review text. The exact method depends on the HTML structure.

```python
for review in reviews:
    # Assuming the review text is inside a <p> tag
    review_text = review.find('p').text
    print(review_text)
```

#Consolidated code example

```python
# Import libraries
from bs4 import BeautifulSoup
import requests

# Make a request
url = 'YOUR_TARGET_URL'
response = requests.get(url)

# Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Locate and extract reviews
reviews = soup.find_all('div', class_='review')
for review in reviews:
    review_text = review.find('p').text
    print(review_text)
```
